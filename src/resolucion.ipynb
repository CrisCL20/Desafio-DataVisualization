En base a los outputs que acabamos de obtener, está claro que "institucion" clasifica por institucion, año, ingresos totales, ingresos fuera de lo académico, ingresos por matrícula, gastos totales, gastos de personal y gastos hacia la facultad (se podría intuir que son gastos de infraestructura). Los datos que encontramos en "program" son clasificados por programa, institución, nivel del programa, clasificación o categoría del programa, duración del programa en años, un identificador que nos dice si el programa está acreditado internacionalmente (1) o no (0), año del programa, estudiantes nuevos que están en el programa, estudiantes mujeres nuevas que están en el programa, estudiantes internacionales nuevos que están dando el programa, estudiantes inscritos, estudiantes internacionales inscritos, estudiantes mujeres inscritas, estudiantes graduados, estudiantes mujeres graduados, estudiantes internacionales graduados y tarifas por matrícula.
En base a esta información, podemos obtener por ejemplo cuántos estudiantes son mujeres (incluyendo las nuevas y las que se estan graduando) y están dando el programa "PROGRAM 1": 

```python
prog = prog_df['program']
fem_students = prog_df[['new_female_students_enrolled','female_students_enrolled','female_graduating_students']]
fem_sum = fem_students.sum(axis=1)
total=0
i=0
for p in prog:
    if p=="PROGRAM 1":
        total+=fem_sum.iloc[i]
    i+=1
print(f"hay {total} mujeres dando el programa \"POGRAM 1\"")
```
