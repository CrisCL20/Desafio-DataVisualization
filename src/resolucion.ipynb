En base a esta información, podemos obtener por ejemplo cuántos estudiantes son mujeres (incluyendo las nuevas y las que se estan graduando) y están dando el programa "PROGRAM 1": 

```python
prog = prog_df['program']
fem_students = prog_df[['new_female_students_enrolled','female_students_enrolled','female_graduating_students']]
fem_sum = fem_students.sum(axis=1)
total=0
i=0

#Codigo mejorable
for p in prog:
    if p=="PROGRAM 1":
        total+=fem_sum.iloc[i]
    i+=1
print(f"hay {total} mujeres dando el programa \"POGRAM 1\"")
```

#### Visualización de los datos de institucion

Primero analizaremos una comparación de ingresos totales en cada institucion en los año 2019 y 2020 (por separado) versus los gastos totales de las instituciones esos mismos años:

```python
temp = ins_df[['institution','year','total_income','total_expenditure']]
df_19 = temp.loc[(temp['year']==2019)]
df_20 = temp.loc[(temp['year']==2020)]

index_19 = df_19['institution'].values.tolist()

df19 = pd.DataFrame({'Ingresos':df_19['total_income'].values.tolist(),
                      'Gastos':df_19['total_expenditure'].values.tolist()},
                      index=index_19)


df19.plot.barh()
plt.legend(bbox_to_anchor=(0,0))
plt.title(label='2019')

index_20 = df_20['institution'].values.tolist()

df20 = pd.DataFrame({'Ingresos':df_20['total_income'].values.tolist(),
                      'Gastos':df_20['total_expenditure'].values.tolist()},
                      index=index_20)


df20.plot.barh()
plt.legend(bbox_to_anchor=(0,0))
plt.title(label='2020')

plt.show()
```

Lo que nos quiere decir esta representación de los datos es que, para el año 2019 el instituto 'U'aún no formulaba en la base de datos entregada y además se puede observar que para ese mismo año las instituciones 'EPSILON','GAMMA','DELTA' tuvieron más gastos que ingresos lo cual puede ser debido a inversiones en infraestructura y personal ó bien una mala gestion de los fondos por parte de la institución. En el año 2020 el instituto 'U' ingresó en la base de datos, con ingresos muchos mayores que el resto de instituciones pero con unos gastos 8 veces mayores. Antes de entrar a analizar que es lo que pasó en ese caso en específico, eliminaremos esa fila en específico para poder hacer un mejor analisis gráfico respecto al resto de los institutos:

```python
index = df_20['institution'].values.tolist()
rem_index = index.index('U')
index.remove('U')

ingresos = df_20['total_income'].values.tolist()
del ingresos[rem_index]

gastos = df_20['total_expenditure'].values.tolist()
del gastos[rem_index]

df20 = pd.DataFrame({'Ingresos':ingresos,
                     'Gastos':gastos},
                     index=index)

df20.plot.barh()
plt.legend(bbox_to_anchor=(0,0))
plt.title(label='2020 modificado')

plt.show()
```

Como se puede ver, los ingresos de las insituciones 'EPSILON','DELTA','GAMMA' son mayores que los gastos totales, mientras que para el resto de instituciones pasó lo contrario con respecto al 2019, es decir, éstas cuentan con más gastos que ingresos y puede ser bien por un mal manejo de recursos ó inversión en otras áreas de la institución.
